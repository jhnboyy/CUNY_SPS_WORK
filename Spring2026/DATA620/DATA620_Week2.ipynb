```python
import networkx as nx
import matplotlib.pyplot as plt
#https://brightspace.cuny.edu/d2l/le/dropbox/1154246/2348830/DownloadAttachment?fid=67923800
```

```python
# Estaplishing an empty graph
G = nx.Graph()
## Adding the nodes in the photo
G.add_node("Andre")
G.add_node("Beverly")
G.add_node("Carol")
G.add_node("Diane")
G.add_node("Ed")
G.add_node("Fernando")
G.add_node("Garth")
G.add_node("Heather")
G.add_node("Ike")
G.add_node("Jane")
```

```python
### Adding the Edges to mimic the diagram in the assignment

## Andrew Edges First 
G.add_edge("Andre", "Beverly")
G.add_edge("Andre", "Carol")
G.add_edge("Andre", "Diane")
G.add_edge("Andre", "Fernando")

## Additional Beverly Edges
G.add_edge("Beverly", "Diane")
G.add_edge("Beverly", "Ed")
G.add_edge("Beverly", "Garth")

## Additional Ed Edges
G.add_edge("Ed", "Garth")
G.add_edge("Ed", "Diane")

## Additional Diane Edges
G.add_edge("Diane", "Garth")
G.add_edge("Diane", "Fernando")
G.add_edge("Diane", "Carol")

## Additonal Carol Edges 
G.add_edge("Carol", "Fernando")

## Additional Garth Edges
G.add_edge("Garth", "Fernando")
G.add_edge("Garth", "Heather")

## Additional Fwernando Edgews 
G.add_edge("Fernando", "Heather")

## Additional Heather Edges 
G.add_edge("Heather",'Ike')

#Jane Edge
G.add_edge("Jane", "Ike")
```

```python
## Diplaying the graph
```

```python
## Printing out
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True)
plt.show()
```
