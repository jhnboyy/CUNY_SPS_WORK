```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom, geom
# Determine how many trials it would take for a fair die to land on the number 4. 

## Parameters 
### Probabiltiyu of Dice landing on 4
p = (1/6)
### Number of Successes we want
k = 1
## Number of trials we want to find probability for 
x = np.arange(0,30)

## Negative Binomial
neg_binomial = nbinom.pmf(x, k, p)

## Geometric Distribution
geometric_prob = geom.pmf(x, p)

# using Demostrative code for plotting
plt.figure(figsize=(8, 5))
plt.bar(x - 0.2, geometric_prob, width=0.4, label="Geometric", color="blue", alpha=0.6)
plt.bar(x + 0.2, neg_binomial, width=0.4, label="Neg Binomial", color="red", alpha=0.6)

# Add labels and legend
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Comparison of Neg. Binomial and Geometric Distributions")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
