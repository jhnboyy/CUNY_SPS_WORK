```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
# Step 1: Define the x1 range for plotting
x1 = np.linspace(-5, 5, 100)  # Generate 100 points from -5 to 5

# Step 2: Define equations from the image, solving for x2 in terms of x1
# Equation 1: 2x1 + 3x2 = 3  ->  Solving for x2: x2 = (3 - 2x1) / 3
eq1 = (3 - 2*x1) / 3  

# Equation 2: x1 - x2 = 4  ->  Solving for x2: x2 = x1 - 4
eq2 = x1 - 4        

# Equation 3: 4x1 + 6x2 = 6  ->  Solving for x2: x2 = (6 - 4x1) / 6
eq3 = (6 - 4*x1) / 6  

# Equation 4: 4x1 + 6x2 = 10  ->  Solving for x2: x2 = (10 - 4x1) / 6
eq4 = (10 - 4*x1) / 6  

# Step 3: Create the plot
plt.figure(figsize=(8,6))  # Set figure size

# Step 4: Plot each equation as a line
plt.plot(x1, eq1, label=r'$2x_1 + 3x_2 = 3$', linestyle='solid', color='b')  # Blue solid line
plt.plot(x1, eq2, label=r'$x_1 - x_2 = 4$', linestyle='dashed', color='r')   # Red dashed line
plt.plot(x1, eq3, label=r'$4x_1 + 6x_2 = 6$', linestyle='dotted', color='g')  # Green dotted line
plt.plot(x1, eq4, label=r'$4x_1 + 6x_2 = 10$', linestyle='dashdot', color='m')  # Magenta dashdot line

# Step 5: Mark the unique solution point (3, -1) for the first system
plt.scatter(3, -1, color='black', marker='o', label="Unique Solution (3, -1)")

# Step 6: Add labels, legend, and grid for better readability
plt.axhline(0, color='black', linewidth=0.5)  # Add horizontal axis
plt.axvline(0, color='black', linewidth=0.5)  # Add vertical axis
plt.xlabel("$x_1$")  # Label x-axis
plt.ylabel("$x_2$")  # Label y-axis
plt.legend()  # Display legend to explain the lines
plt.title("Graphical Representation of Three Systems of Linear Equations")  # Title for the plot

# Step 7: Display the grid and the final plot
plt.grid()  # Enable grid for easier visualization
plt.show()  # Render the plot
```

```python
#  Exercise LC.C40
### Importing potentially relevant libraries 
import sympy as sp #(Found this via google for Math related work/notation in python)

#### Firstly, defininign the variables ot be used in the problem
x1, x2, x3, x4, x5 = sp.symbols('x1 x2 x3 x4 x5')
####  Outlining the original equations
eq1 = sp.Eq(2*x1 - 4*x2 + 3*x3 + x5, 6)
print(eq1)
eq2 = sp.Eq(x1 - 2*x2 - 2*x3 + 14*x4 - 4*x5, 15)
print(eq2)
eq3 = sp.Eq(x1 - 2*x2 + x3 + 2*x4 + x5, -1)
print(eq3)
eq4 = sp.Eq(-2*x1 + 4*x2 - 12*x4 + x5, -7)
print(eq4)

#### Placing these  equations into vector form: a
v_eq1 = [2,-4,3,0,1,6]
v_eq2 = [1,-2,-2,14,-4,15]
v_eq3 = [1,-2,1,2,1,-1]
v_eq4 = [-2,4,0,-12,1,-7]

#### Placing the Vectors into a matrix 
mtx = sp.Matrix([v_eq1,v_eq2,v_eq3,v_eq4])
print(mtx)

#### Completing the RREF Transform
rref_mtx = mtx.rref()
print(rref_mtx)
#Note Columns 2 and 4 have no 1 values. This implied that x2 and x4 are the free variables. 
#The rest (x1,x3,x5) are dependent.
#Due to the presense of free variables there are infinite solutions. 
    
### The next step is expressing the equations in terms of free variables (x2 and x4)
### Using augmented matrix
x1 - 2x2 -6x4 = 1 -> x1 = 1 + 2x2 +6x4
x3 - 4x4 = 3 -> x3 = 3 + 4x4
x5 = -5

### Writing them as vectors
[x1,x2,x3,x4,x5] = [1,0,3,0,-5] + x2[2,1,0,0,0] + x4[6,0,4,1,0]

```
