```python
import numpy as np
import matplotlib.pyplot as plt

## Using Non zero values because of the formulat, dont want to divide by zero
x_values = np.linspace(0.1,0.9,100)
# print(x_values)

## Removed factrorial from demo function because its a geometric series and the facvtorials cancel
def maclaurin(x, terms):
    series = sum([x**n for n in range(terms)])
    return series

def number5(x):
    val = 1/(1-x)
    return val

#Getting the real values from calculating it: 
actual = [number5(x) for x in x_values]
est_val = [maclaurin(x, 100) for x in x_values]

## Plotting 
plt.plot(x_values, actual, color="blue")
plt.plot(x_values, est_val, color="red", linestyle="--")
```
