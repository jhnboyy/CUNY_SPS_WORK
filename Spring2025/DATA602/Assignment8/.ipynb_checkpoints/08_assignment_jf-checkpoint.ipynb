# Data Wrangling

Perform data wrangling.  You are free to use your best judgment here.  If you are stuck, look at previous assignment.

```python
# Aggregating to get numbers for analysis and plotting 
df_agg = df.groupby(["Area Name","Area ID","Primary Fur Color",
            "Highlights in Fur Color","Location","Activities","Interactions with Humans"
           ]).agg({"Squirrel ID":'nunique'}).reset_index()
df_agg = df_agg.rename(columns={"Squirrel ID":"Squirrel_Count"})
```

#### Additional Exploration

```python
## Primary Color Totals 
color_agg= df_agg.groupby(["Primary Fur Color"]).agg({'Squirrel_Count':'sum'}).reset_index()
## ----- Numeric Summary
print(color_agg)

## Only three rows with the vast majority of the squirrels being gray as their primary color.

## Location in park totals
location_agg= df_agg.groupby(["Location"]).agg({'Squirrel_Count':'sum'}).reset_index()
## ----- Numeric Summary
print(location_agg.describe())

## Notes
#There are 6 different locations where squirrels were noted to be found withing parks, the mean across the areas is about 45 squirrels. 
#The max amount for a speicifc location is 169, while the minimum is 5. 
```

```python
## Aggregated Level 1 in orderto get counts for analysis.

## Simplifying and cleaning 'Interactions with Humans' Column
## Categorizing the interactions with Humans Column for simplicity. 
## Categories: Cautiously Curious, Human Friendly, Indifferent, Skittish.
df_agg["HumanRelationship"] = None
## Cautiously Curious
df_agg["HumanRelationship"][df_agg['Interactions with Humans'].isin(['Staring',
                                                                    'Approaches, Runs From','Watching',
                                                                     'Watches us from tree','Watching us from tree - very interested in us'
                                                                    ])]="Cautiously Curious"
## Human Friendly
df_agg["HumanRelationship"][df_agg['Interactions with Humans'].isin(['Approaches','Friendly',
                                                                    'Approaches, watching us',
                                                                    'Okay with people'])]="Human Friendly"
## Indifferent
df_agg["HumanRelationship"][df_agg['Interactions with Humans'].isin(["Indifferent",
                                                                     'Indifferent, Preoccupied by HAAWK',
                                                                     'Indifferent, Runs From'])]="Indifferent"
## Skittish
df_agg["HumanRelationship"][df_agg['Interactions with Humans'].isin(['Runs From',
                                                                    'Skittish to humans',
                                                                     'Runs From, watches us in short tree',
                                                                     'Runs From, watchful'
                                                                    ])]="Skittish"
#df_agg['Interactions with Humans'].unique()
```

```python
## First Visualization Set
## Plotting Primary Color By Area 
area_color = df_agg.groupby(["Area Name","Primary Fur Color"]).agg({"Squirrel_Count":"sum"}).reset_index()
for area in area_color['Area Name'].unique():
    area_temp = area_color[area_color['Area Name']==area].sort_values(["Squirrel_Count"],ascending=False)
    bar_colors = area_temp['Primary Fur Color'].map({'Gray': 'gray','Cinnamon': 'brown','Black': 'black'})
    plt.bar(area_temp['Primary Fur Color'],area_temp["Squirrel_Count"], color=bar_colors)
    plt.title(f"Count of Squirrels in {area.title()} by Primary Fur color")
#     plt.xlabel('Primary Color') ## Think it looks better without. 
    plt.ylabel("Count of Squirrels",fontsize=12)
    plt.xticks(rotation=45,ha='right',fontsize=12)
    plt.show()
```

```python
print(df_agg["HumanRelationship"].unique())
```

```python
## Second Visualization Set 
## Plotting HumanRelationship by area
area_rel = df_agg.groupby(["Area Name","HumanRelationship"]).agg({"Squirrel_Count":"sum"}).reset_index()
for area in area_rel['Area Name'].unique():
    area_temp = area_rel[area_rel['Area Name']==area].sort_values(["Squirrel_Count"],ascending=False)
    bar_colors = area_temp['HumanRelationship'].map({'Indifferent':"yellow", 'Skittish':"red", 'Cautiously Curious':"orange", 'Human Friendly':"green"})
    plt.bar(area_temp['HumanRelationship'],area_temp["Squirrel_Count"], color=bar_colors)
    plt.title(f"Count of Squirrels in {area.title()} by Human Receptivity")
#     plt.xlabel('Primary Color') ## Think it looks better without. 
    plt.ylabel("Count of Squirrels",fontsize=12)
    plt.xticks(rotation=45,ha='right',fontsize=12)
    plt.show()
```
