```python
import pandas as pd
df = pd.read_csv("squirrel-data.csv",encoding="ISO-8859-1")## Shifting encoding b.c default UTF-8 breaks
##
```

```python
## Takign a look at the head 
print(df.columns)
df.head()
```

```python
### ---- Notes
# This is geolocated data so there are multiple columns which we are most likely not going to use for this base level analysis. These columns include: 
# Squirrel Latitude (DD.DDDDDD)
# Squirrel Longitude (-DD.DDDDDD)

## Seeing what describe shows.
# print(df.describe())

# No numeric data in here except for an identifier and the location coordinates. Need to aggregate to get counts to work with. 
# This data is mainly qualitative data, with the exception of the lat and long.
# This means we will need to aggregate row counts in order to get some level of numeric content. 
#
### ---- For the remaining columms that are not numeric, which are the columns we have interest in leveraging in aggregate analysis. 
for col in df.columns: 
    print('-------------------------------------------')
    print(f"Column: {col}")
    print(f'Unique Values for {col}: ',df[col].unique())
    print(f"Null values for {col}: ",df[col].isnull().sum())
    print('-------------------------------------------')
    
```

```python
## Looking at the break down for the different columns and thier values I think aggregating up to the "Area Name", 
## the part of the city documented, along with the squirrel information.


### Taking note of the nulls present in the data for those columns where nulls are present and relevant.Filling them in with an "NA"
    #Null values for Primary Fur Color:  1
    #Null values for Highlights in Fur Color:  94   
    #Null values for Color Notes:  423
    #Null values for Location:  34
    #Null values for Above Ground (Height in Feet):  321
    #Null values for Specific Location:  344
    #Null values for Activities:  55
    #Null values for Interactions with Humans:  91
    #Null values for Other Notes or Observations:  266
```

```python
# Aggregating to get numbers
df_agg = df.groupby(["Area Name","Area ID","Primary Fur Color",
            "Highlights in Fur Color","Location","Activities","Interactions with Humans"
           ]).agg({"Squirrel ID":'nunique'}).reset_index()
df_agg = df_agg.rename(columns={"Squirrel ID":"Squirrel_Count"})
```

```python
## Area Name totals
area_agg = df_agg.groupby(["Area Name","Area ID"]).agg({'Squirrel_Count':'sum'}).reset_index()
print(area_agg)
## ----- Numeric Summary
area_agg.describe()

## --- Notes: 
#Only 4 areas in the data, the mean for each area is aroudn 67 squirrels. 
#The minimum number of squirrels in an area is 30 and the max is 108. 
```

```python
## Primary Color Totals 
color_agg= df_agg.groupby(["Primary Fur Color"]).agg({'Squirrel_Count':'sum'}).reset_index()
## ----- Numeric Summary
print(color_agg)

## Only three rows with the vast majority of the squirrels being gray as their primary color.
```

```python
## Location in park totals
location_agg= df_agg.groupby(["Location"]).agg({'Squirrel_Count':'sum'}).reset_index()
## ----- Numeric Summary
location_agg.describe()

## Notes
#There are 6 different locations where squirrels were noted to be found withing parks, the mean across the areas is about 45 squirrels. 
#The max amount for a speicifc location is 169, while the minimum is 5. 
```

```python
## Activites Totals. 
activity_agg = df_agg.groupby(["Activities"]).agg({'Squirrel_Count':'sum'}).reset_index()
## ----- Numeric Summary
activity_agg.describe()
## Notes:
##  The activies column will need cleaning, Although for now the numeric summary is as follows:
## There are 48 different unprocessed activity categories, with an average number of squirrels for each at about 5.6. 
# The activity with the largest number of squirrels associated with is has a max of 78 squirrels categorized with the activity,
# while the minimum activity has 1 squirrel. These categories need to be cleaned if being analyzed. 
```
