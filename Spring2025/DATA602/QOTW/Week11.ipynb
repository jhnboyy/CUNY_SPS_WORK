```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sn
import kagglehub
```

```python
## Reading in example data. 
path = kagglehub.dataset_download("yasserh/student-marks-dataset")
home_dir = os.path.expanduser("~")
file_dir = os.path.join(home_dir,path,"Student_Marks.csv")
df = pd.read_csv(file_dir)
print(df.head())
```

#### 1) How do you plot a histogram in Seaborn?

```python
#Making the histogram
print(sn.histplot(data=df, x="Marks",color=sn.color_palette("hls", 10)[3]))
```

#### 2) Plot a histogram with NAs dropped.

```python
## my Df has no nulls, so will add some to drop via seaborn
na_data = pd.DataFrame({
    'number_courses': [5, 3, 6],
    'time_study': [np.nan,np.nan,np.nan],
    'Marks': [25.4, 30.6, 45.2]
})
df_w_na = pd.concat([na_data,df])

#Plitting histogram with NAs dropped
print(sn.histplot(data=df_w_na, x="time_study",color=sn.color_palette("hls", 10)[5]))

## Seaborn automatically ignores nulls. 
```

#### 3) How do you set the color for a histogram?

```python
# Already doing this with the color input. 
# I pull up a pallet via name, and the number of colors i want. I then select via list index.
print(sn.histplot(data=df_w_na, x="time_study",color=sn.color_palette("hls", 10)[2]))
```

#### 4) What type of plot would allow you to compare two continuous features? Give an example of code.

```python
# A scatter plot could show two continuous variables. Lets do marks and time study
print(sn.scatterplot(data=df_w_na, x='time_study', y='Marks',color=sn.color_palette("hls", 10)[4]))
```

#### 5) Give example of a correlation plot.

```python
# First i need a correlation matrix using the data.
cor_matx = df.corr()
print(sn.heatmap(cor_matx))
```

#### 6) Change the figure size of your plot(s). You can use any dataset for examples to these questions. 

```python
## Changing figure size with matplotlib
plt.figure(figsize=(8, 8))
sn.heatmap(cor_matx)
plt.show()
```
